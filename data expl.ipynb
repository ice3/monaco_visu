```python
def get_quartiers(data):
    for q in data['Document']['Quartier']:
        yield q
```

```python
def get_nom_quartiers(data):
    for q in get_quartiers(data):
        yield q['nomQuartier']
```

```python
for q in get_nom_quartiers(DATA):
    print(q)
```

```python
def nb_free_spots_quartier(data):
    for q in get_quartiers(data):
        yield q['totalLibresQuartier']
```

```python
for q in nb_free_spots_quartier(DATA):
    print(q)
```

```python
def get_all_parcs(data):
    for q in get_quartiers(data):
        for p in q['Parc']:
            if isinstance(p, dict):  # some are unicde, idk what it is
                yield p
```

```python
for p in list(get_all_parcs(DATA))[:10]:
        print(p['libelleParc'])
```

```python
def get_places_parcs(data):
    for p in get_all_parcs(data):
        try:
            di = {'nom': p['libelleParc'],
                  'libres': p['placesLibresParc'],
                  'occupes': p['presents']['presentsTotal']}
            if di['libres'] == u'Obsolete':
                print(di, "OBSOLETE OBSOLETE OBSOLETE OBSOLETE OBSOLETE OBSOLETE OBSOLETE OBSOLETE OBSOLETE OBSOLETE OBSOLETE OBSOLETE OBSOLETE OBSOLETE OBSOLETE OBSOLETE ")
            yield di
        except ValueError:
            print(di, "OBSOLETE OBSOLETE OBSOLETE OBSOLETE OBSOLETE OBSOLETE OBSOLETE OBSOLETE OBSOLETE OBSOLETE OBSOLETE OBSOLETE OBSOLETE OBSOLETE OBSOLETE OBSOLETE ")
```

```python
for pl in get_places_parcs(DATA):
    print(pl)
```

```python
pl_dict = {pl['nom']: int(pl['libres']) + int(pl['occupes']) for pl in get_places_parcs(DATA)}
```

```python
sorted_keys = sorted(pl_dict, key=pl_dict.get, reverse=True)
```

```python
import glob
import os
import datetime
```

```python
DATAS = {}
fnames = glob.glob('data_json/samples/json/2014/04/16/*.json')
for fname in fnames:
    with open(fname, 'r') as f:
        data = json.load(f)
    h = os.path.split(fname)[-1]
    h = os.path.splitext(h)[0]
    h = datetime.datetime.strptime(h, '%H_%M')
    DATAS[h] = data
```

```python
x_heure = []
y_pl_libres  =[]
for dtime, d in DATAS.items():
    x_heure.append(dtime)
    y_pl_libres.append(list(get_places_parcs(d))[0]['libres'])
    print(list(get_places_parcs(d))[0]['libres'], list(get_places_parcs(d))[0])
```
